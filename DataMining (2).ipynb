# Praproses Data Komoditas Pasar Sederhana Tahun 2017

Pada Notebook ini dijelaskan mengenai praproses dari beberapa data harga komoditas pasar Sederhana tahun 2017. Langkah-langkah yang dilakukan antara lain:
1. Memuat data .csv dari bulan Januari hingga Desember ke dalam 12 Variabel
2. Mentransposkan data
3. Menghilangkan kolom "lokasi" pada bulan pertama
4. Menghilangkan kolom "nama_komoditas" pada bulan kedua hingga bulan terakhir
5. Menyatukan semua variabel (concat)

```python
frames = [dfSederhanaJan, dfSederhanaFeb, dfSederhanaMar, dfSederhanaApr, dfSederhanaMei, dfSederhanaJun, dfSederhanaJul, dfSederhanaAgt, dfSederhanaSep, dfSederhanaOkt, dfSederhanaNov, dfSederhanaDes]
result = pd.concat(frames)
result
```

### Menangani Beberapa Baris yang Berisikan Nilai 0 Seluruhnya

Terdapat beberapa baris yang isinya merupakan 0, maka akan dihilangkan.

```python
# Transposekan terlebih dahulu, lalu filter
result = result[(result.T != 0).any()]
# Tampilkan
result
```

```python
means = result.copy()
means = means.T
means = means.infer_objects().fillna(0.0)
means['rata-rata'] = means.mean(numeric_only=True, axis=1)
means.T
```
